```python
def quick_calc(input_df):
    df = input_df.copy()
    res_dict = {}
    for player in df.index:
        abstand = df.loc[player,'total_points'] - df.total_points
        pos_abstand = abstand > 0
        counter = 0
        if sum(pos_abstand) > 0:
            reduced_point_list = [x for x in points_per_position if x > abstand[pos_abstand].iloc[counter]]
            while True:
                reduced_point_list.pop(-1)
                counter += 1
                try:
                    reduced_point_list = [x for x in reduced_point_list if x > abstand[pos_abstand].iloc[counter]]
                    if len(reduced_point_list) == 0:
                        break
                except:
                    pass
            res_dict[player] = min(128,df.loc[player,'ranking'] + counter)
        else:
            res_dict[player] = 128
        df['min_rank'] = df.index.map(res_dict)
    return df
```

```python
res = quick_calc(df)
```
